```python
import pandas as pd
url = "https://drive.google.com/uc?export=download&id=144dlnCJ7rHs4LllVqXtsg2ABpnN_XucB"
df = pd.read_csv(url)

df = df[['Where are you based?','Which age range are you in?', 'How would you describe your gender?', 'Are you living with a disability or a long-term condition? If so how would you describe this?']]

def encode_gender(df, **kwds):

      key = kwds['key']

      if df[key] == "Female":
        return df[key]
      if df[key] == "Male":
        return df[key]
      if df[key] == "Prefer not to say":
        return df[key]
      else:
        return "Other"


df["gender-encoded"] = df.apply(encode_gender, axis=1, key="How would you describe your gender?")
df

#Testing the count:

# def test_count_gender():
#   print(df['gender-encoded'].value_counts())
# test_count_gender()
```
