```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://drive.google.com/uc?export=download&id=144dlnCJ7rHs4LllVqXtsg2ABpnN_XucB"
df = pd.read_csv(url)

df = df[['Where are you based?', 'Which age range are you in?', 'How would you describe your gender?', 'Are you living with a disability or a long-term condition? If so how would you describe this?']]

def lowercase():
    df['Are you living with a disability or a long-term condition? If so how would you describe this?'] = df['Are you living with a disability or a long-term condition? If so how would you describe this?'].str.lower()
lowercase()

def encode_disability(df, **kwds):

    key = kwds['key']

    if df[key] == "no":
        return 1
    else:
        return 0

df["disability-encoded"] = df.apply(encode_disability, axis=1, key="Are you living with a disability or a long-term condition? If so how would you describe this?")

# Python function 1 run once for the Disability column with pie chart fully labelled and downloaded as .png

def produce_piechart(df, column):

    data = df[column]
 #   labels = data.value_counts().keys()
    colors = ['#ebf5df', '#bad4aa']
    labels = ["Non-disabled", "Disabled"]
    plt.figure(figsize=(10,10))
    plt.pie(data.value_counts(), colors = colors, autopct='%1.1f%%', wedgeprops={'edgecolor':'white'})
    plt.legend(labels=labels, bbox_to_anchor=(1.1,0.5), loc="center right", fontsize=15, bbox_transform=plt.gcf().transFigure)
    plt.title("Percentage of Disabled and Non-disabled Participants", fontsize=15)
    plt.show()

produce_piechart(df, "disability-encoded")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://drive.google.com/uc?export=download&id=144dlnCJ7rHs4LllVqXtsg2ABpnN_XucB"
df = pd.read_csv(url)

df = df[['Where are you based?','Which age range are you in?', 'How would you describe your gender?', 'What is your ethnicity?', 'Are you living with a disability or a long-term condition? If so how would you describe this?']]

# Python function 1 run once for the Ethnicity column with pie chart fully labelled and downloaded as .png

def produce_piechart(df, column):
    colors = ['#BAEBFF', '#BBDBFE', '#BCCBFD', '#BEBCFC', '#BFACFB', '#C09CFA', '#C18CF9',
              '#C37DF8', '#C46DF7', '#C55DF6', '#A564D3', '#B66EE8', '#C879FF', '#D689FF', '#E498FF', '#F2A8FF', '#FFB7FF',
              '#FFC4FF', '#FFC9FF', '#FFCEFF', '#fed7fa', '#FCD6F2', '#FCC6EE', '#FDB7E9', '#FEA7E4', '#FE98E0', '#FF88DB',
              ]

    data = df[column]
    labels = data.value_counts().keys()
    plt.figure(figsize=(10,10))
    plt.pie(data.value_counts(), colors = colors, autopct='%1.1f%%', wedgeprops={'edgecolor':'white'}, pctdistance=1.2, labeldistance=2.0)
    plt.legend(labels=labels, bbox_to_anchor=(1.4,0.5), loc="center right", fontsize=15, bbox_transform=plt.gcf().transFigure)
    plt.title("Percentage of Each Ethnicity", fontsize=15)
    plt.show()

produce_piechart(df, 'What is your ethnicity?')

```
